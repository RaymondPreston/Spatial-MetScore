# Advanced Multi-Reference Spatial Deconvolution with Site-Specific Bolstering

**Goal:** Achieve high-confidence cell type deconvolution of Visium spots by integrating site-specific bolstered references to account for the extreme scarcity of PDAC Lung and Peritoneal scRNA-seq data.

## The Bolstered Reference Strategy

### Layer 1: Site-Specific Reference Bolstering
To prevent "Primary-bias," we supplement the PDAC atlas with high-quality references from other sources to capture the background tissue signatures of the metastatic sites:
- **Primary:** Covered extremely well
- **Liver:** Covered extremely well
- **Lung:** Integration of the **Human Lung Cell Atlas (HLCA)** for healthy baseline, AND maybe from primary lung cancer (NSCLC) and other lung metastasis datasets.
- **Peritoneal :** Integration of normal peritoneum and any other available metastatic peritoneum scRNAseq data

### Layer 2: Balanced Downsampling
We will perform **stratified downsampling** on the overrepresented Primary PDAC and Liver Met cells. This ensures that the rare Lung and Peritoneal signatures have equal "weight" in the Cell2location training model.

### Layer 3: Site-Aware Mapping
We will apply the deconvolution in a "Site-Aware" manner:
- Lung Visium slices will be mapped using the Lung-Bolstered reference.
- Peritoneal Visium slices will be mapped using the Peritoneal-Bolstered reference.
